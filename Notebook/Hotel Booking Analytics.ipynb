```python
df["Arrival_Date"] = pd.to_datetime(
    df["arrival_date_day_of_month"].astype(str) + "-" +
    df["arrival_date_month"].astype(str) + "-" +
    df["arrival_date_year"].astype(str),
    format="%d-%B-%Y"
)
```

```python
df.head()
```

```python
df["Total_Guest"] = df["adults"] + df["children"] + df["babies"]
```

```python
df.head()
```

```python
df.drop(columns=["arrival_date_year", "arrival_date_day_of_month", "arrival_date_month" ], inplace=True)
```

```python
df.head()
```

```python
df["cust_satisfication"] = np.where(
    df["reserved_room_type"] == df["assigned_room_type"],
    "Satisfied",
    "Not Satisfied"
)
```

## Handling null values

```python
df.isnull().sum()
```

```python
df["country"] = df["country"].fillna("UD")
```

```python
df.dropna(subset=["Total_Guest"], inplace=True)
```

```python

# Chart - 2 visualization code
# group by hotel
grup_by_hotel=df.groupby('hotel')


#grouping by hotel adr
highest_adr=grup_by_hotel['adr'].mean().reset_index()

#set plot size
plt.figure(figsize=(11,9))

#plot the graph
sns.barplot(x=highest_adr['hotel'],y=highest_adr['adr'])

# set labels
plt.xlabel('Hotel type')
plt.ylabel('ADR')
plt.title("Avg ADR of each Hotel type")
```

```python
# Chart - 1 visualization code
hotel_value_counts = df['hotel'].value_counts()
hotel_value_counts



#Pie Chart for Most Preffered  Hotel
hotel_value_counts.plot.pie(figsize=(11,9),shadow=True,fontsize=20, autopct='%1.2f%%',)   

plt.title('Most Preffered  Hotel',fontsize = 10)
plt.show()
     
```

```python
# Chart - 1 visualization code
hotel_value_counts = df['cust_satisfication'].value_counts()
hotel_value_counts



#Pie Chart for Most Preffered  Hotel
plt.figure(figsize=(11,9))

hotel_value_counts.plot(
    kind="bar",
    color=["#4CAF50", "#F44336"],
    fontsize=15
)

plt.title("Customer Satisfaction Count", fontsize=16)
plt.xlabel("Satisfaction")
plt.ylabel("Count")
plt.show()
     
```
